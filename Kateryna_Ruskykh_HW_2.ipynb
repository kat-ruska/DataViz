```
import pandas as pd
import altair as alt
```

```
alt.data_transformers.enable('default', max_rows=None) 
```

```
df = pd.read_csv('https://raw.githubusercontent.com/kat-ruska/DataViz/master/HW2/rate.csv')
print(df.head())
```

```
chart = alt.Chart(df, title="Population growth rate by region").mark_rect().encode(
      x = alt.X(
          'year:O',
          title = None,
          scale = alt.Scale(zero = False),
          axis=alt.Axis(format = 'd', 
                        values = [1990, 1995, 2000, 2005, 2010, 2015],
                        orient = 'top',
                        labelAngle = 0,
                        ticks = False,
                        labelFontSize = 12,
                        domainColor = 'white',
                        labelColor = '#5D646F',
                        labelFont ='Ubuntu Mono')
      ),
      y = alt.Y('region:N',
                title = None,
                scale = alt.Scale(zero = False),
                sort=alt.Sort(field = 'rate', order='descending'), 
                axis=alt.Axis(ticks = False,  
                              orient = 'right',
                              #domain = False, 
                              domainColor = 'white',
                              labelFontSize = 12,
                              #grid = False,
                              labelColor = '#5D646F',
                              labelFont ='Ubuntu Mono'
                              )
                ), 
      color = alt.Color('rate',
                        scale=alt.Scale(domain = [-15, 0, 8],
                                        range=['#67001f', '#f7f7f7', '#053061']),
                        legend=alt.Legend(orient="top",
                                          title = 'population growth rate, per 100',
                                          gradientLength = 799,
                                          gradientThickness = 10,
                                          labelColor = '#5D646F',
                                          labelFontSize = 12,
                                          titleColor = '#5D646F',
                                          titleFontWeight = 'normal',
                                          titleFontSize = 12,
                                          titleFont ='Ubuntu Mono',
                                          labelFont ='Ubuntu Mono',
                                          labelAlign = 'left',
                                          values = [-15, -8, 0, 8]))
).properties(width = 800)
             
chart
```

```
caption = alt.Chart(
    {"values": [{"text": "Data: State Statistics Service of Ukraine"}]}
).mark_text(
    size=12,
    font = 'Ubuntu Mono',
    align = 'left',
    color = '#5D646F',
    baseline = 'bottom',
    dx = 200,
    dy = 560

).encode(
    text="text:N",
    y= alt.value(2)
).properties(
    width = 800
)
```

```
alt.layer(chart, caption).configure_title(anchor = 'start',
                                               font = 'Ubuntu Mono',
                                               fontSize = 40,
                                               fontWeight = 'normal',
                                               offset = 15
).properties(background = '#F3F7F7',
             width = 800)
```
