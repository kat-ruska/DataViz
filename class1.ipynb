```python
import pandas as pd
```

```python
data = pd.read_csv("christmas.csv")
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(16,8))
# Fixing random state for reproducibility

x = data['gdpPerCap']
y = data['lifeExp']

xnorm= (x-min(x))*500/(max(x)-min(x))
area = xnorm # 0 to 15 point radii


#colors = np.random.rand(5)

plt.scatter(x, y, s=area, alpha=0.5)

plt.xlabel('GDP per capita')

plt.ylabel('Life Expectancy')

plt.title('Are the GDP per capita and Life Expectancy related?')

plt.show()

```
